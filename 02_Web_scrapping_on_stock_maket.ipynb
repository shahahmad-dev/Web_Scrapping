**Apple stock price data Web_scrapping**


```python
# Import required libraries
import yfinance as yf   # For downloading stock market data
import pandas as pd     # For data manipulation

# Step 1: Define the stock ticker symbol
stock_symbol = 'AAPL'   # AAPL is the ticker symbol for Apple Inc.

# Step 2: Create a stock object using yfinance
stock_object = yf.Ticker(stock_symbol)

# Step 3: Download historical stock data
# We are getting daily stock data from Jan 1, 2010 to Jan 25, 2020
stock_price_data = stock_object.history(
    start='2010-01-01',
    end='2020-01-25'
)

# Step 4: Get the last available closing price in the dataset
last_closing_price = stock_price_data['Close'].iloc[-1]

print("Last Closing Price:", last_closing_price)

# Step 5: Display the first 5 rows of the dataset
stock_price_data.head()
```

**Taking last one year data**

```python
import pandas as pd
import numpy as np
import yfinance as yf
import datetime as dt
from datetime import date, timedelta

today = date.today()

d1 = today.strftime('%Y-%m-%d')

d2 = (today - timedelta(days=365)).strftime('%Y-%m-%d')

start_time = d2
end_time = d1


stock_symbol = 'GOOG'

stock_object = yt.Ticker(stock_symbol)

stock_price_data = stock_object.history(start=start_time, end=end_time)

stock_price_data.head()
```
